```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create

# Read in the data
df = pd.read_csv(r'C:\Users\mwc11\Downloads\moviesdata.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for columns

df.dtypes
```

```python
# change how data type presents (knock off the zeros at the end)

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# create correct year column

df['yearcorrect'] = df['released'].astype(str).str[:4]

df
```

```python
# order by gross revenue

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# set up scrollable option for viewing table

pd.set_option('display.max_rows', None)
```

```python
# drop any duplicates in company column

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#scatter plot with budget vs gross revenue

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show
```

```python
# Plot budget vs gross revenue using seaborn

sns.regplot(x='budget', y='gross', data=df)

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"colour": "blue"})
```
